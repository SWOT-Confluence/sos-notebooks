```python
from netCDF4 import Dataset
from pathlib import Path
import numpy as np
import os
```

```python
SOS=Path('C:/Users/coss.31/OneDrive - The Ohio State University/Documents/SWOT_Mission_REPOS/SOSV2/FinalSoS/ZipHold/unconstrained')
```

```python
FD=os.listdir(SOS)
```

```python
GRID=[]
for F in FD:
    D=Dataset(SOS.joinpath(F))
    AGgrp=D.getncattr('Gage_Agency')
    GP=AGgrp.split(";")
    for g in GP:
        RID=D[g][g+'_reach_id'][:].filled(np.nan)
        GRID.append(RID)
```

```python
OUT=np.concatenate(GRID)
```

```python
import json

with open('GageRIDsGLOBAL.txt', 'w') as filehandle:
    json.dump(list(OUT.astype(str)), filehandle)
```
